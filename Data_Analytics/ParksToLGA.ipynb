```python
import pandas as pd 

# load the recreation data into a data frame.
recs = pd.read_csv("vals .csv")
```

```python
# get a list of different types of rec areas in dataset. 
recs['FEATURETYP'].value_counts().index

# only keep parks and botanic gardens in dataset. 
recs_2 = recs[recs['FEATURETYP'] == 'Park']
recs_3 = recs[recs['FEATURETYP'] == 'Botanic Gardens']

recs_fin = pd.concat([recs_2, recs_3], axis=0)

# remove all other columns 
recs_fin = recs_fin.iloc[:, : 6]

# keep average x,y coordinates of the park or botanic garden. 
recs_fin3 = recs_fin.groupby(['NAME']).mean()
recs_fin2 = recs_fin3.reset_index()
```

```python
# now, load the council areas. 
lga = pd.read_csv("vertex4.csv")
```

```python
# get a list of all council areas. 
lga['ADMINAREAN'].value_counts().index

```

```python
dense_lga = ['BRISBANE CITY', 'MORETON BAY REGIONAL',
       'SUNSHINE COAST REGIONAL',
       'GOLD COAST CITY',
       'SOMERSET REGIONAL',  'CAIRNS REGIONAL',
       'CHARTERS TOWERS REGIONAL',
       'BANANA SHIRE', 'TOOWOOMBA REGIONAL', 'NOOSA SHIRE',
       'LOGAN CITY', 'REDLAND CITY',
       'GLADSTONE REGIONAL', 'DOUGLAS SHIRE',
       'LOCKYER VALLEY REGIONAL', 'IPSWICH CITY', 'TOWNSVILLE CITY']
# remove entries that are not council areas. 

lga = lga[lga['ADMINAREAN'].isin(dense_lga)]

# only keep first five columns 
lga = lga.iloc[:, : 5]

# keep the minimum and maximum coordinates 

lga_max = lga.groupby(['ADMINAREAN']).max()
lga_min = lga.groupby(['ADMINAREAN']).min()
lga_fin = pd.concat([lga_min, lga_max], axis=0)
```

```python

# dictionary with keys
lga_parks = {el:[] for el in dense_lga}


# get dataframes grouped by the range. 
# and just iterate through these mini dataframes instead. 

lga_dfs = []



for d in dense_lga:
    df = lga_fin[lga_fin.index == d]

    lga_dfs.append(df)



```

```python

import math

for index, row in recs_fin2.iterrows():
  row_x = math.floor(row['X'])
  
  for dfs in lga_dfs:

    max = dfs.iloc[0]
    min = dfs.iloc[1]

    if row['X'] <= (max['X'] + 1) and row['X'] >= (min['X']-1) :
      if row['Y'] <= (max['Y'] + 1) and row['Y'] >= (min['Y'] - 1):
          lga_parks[max.name].append(row['NAME'])
    
print(lga_parks)

```
