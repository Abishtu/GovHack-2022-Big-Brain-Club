## Wage Growth Data

```python
df_wage_change = df_wage.iloc[9:,[0,19,20,21]]
df_wage_change
```

```python
df_wage_change = df_wage_change.rename(columns={"Unnamed: 0": "Quarter", df_wage_change.columns[1]: "Private Change",
                                               df_wage_change.columns[2]: "Public Change", df_wage_change.columns[3]
                                               :"Both Change"})
df_wage_change['Quarter'] =  pd.to_datetime(df_wage_change['Quarter'])
df_wage_change.info()
df_wage_change['Public Change'] = pd.to_numeric(df_wage_change['Public Change'])
df_wage_change['Both Change'] = pd.to_numeric(df_wage_change['Both Change'])
df_wage_change['Private Change'] = pd.to_numeric(df_wage_change['Private Change'])
```

```python
plt.figure(1)
plt.plot(df_wage_change['Quarter'], df_wage_change['Both Change'])
```
